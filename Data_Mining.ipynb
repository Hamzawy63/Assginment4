#  Lab1: Data Exploration



### Important Imports & Exploring the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
names = ['ID', 'Diagnosis', 'radius','texture','perimeter','area','smoothness','compactness','concavity','concave points','symmetry','fractal dimension',
'SE radius','SE texture','SE perimeter','SE area','SE smoothness','SE compactness','SE concavity','SE concave points','SE symmetry','SE fractal dimension',
'W radius','W texture','W perimeter','W area','W smoothness','W compactness','W concavity','W concave points','W symmetry','W fractal dimension']

```

```python
data = pd.read_csv("wdbc.data" ,names = names ,  header = None )
print(data)
```

### Column infromation and data type

```python
data.info()
```

### Bie and bar Chart of the diagnosis Column

```python
malignant = data.apply(lambda x: True if x[1] == 'M' else False , axis=1)
print(type(malignant))
number_of_malignant = len(malignant[malignant == True].index)
number_of_benign =len(malignant[malignant == False].index)
print(number_of_malignant)
y = np.array([number_of_malignant, number_of_benign])
mylabels = ["Malignant", "Benign"]

plt.pie(y, labels = mylabels)
plt.show() 



fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['Malignant', 'Benign' ]
students = [number_of_malignant,number_of_benign]
ax.bar(langs,students)
plt.show()
```

### Histograms of the 10 Features 

```python
#Histograms
for i in range(2 , 12):
  data.hist(column = names[i], by = 'Diagnosis', legend = True)
```

### Histogram of the Standard Error

```python
print("Standard Error(SE)")
for i in range(12 , 22):
  data.hist(column = names[i], by = 'Diagnosis', legend = True)
```

### Histogram of worst Features

```python
print("Worst Features") 
for i in range(22 , 32):
  data.hist(column = names[i], by = 'Diagnosis', legend = True)
```
